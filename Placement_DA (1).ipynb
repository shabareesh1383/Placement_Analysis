```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset

df = pd.read_csv('placement_data.csv')

# Display basic information about the dataset
df.info(), df.head()
```

```python
sns.histplot(df["CGPA"], bins=20, kde=True)
plt.title("Distribution of CGPA")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x="PlacementStatus", data=df, palette="coolwarm")
plt.title("Placement Status Distribution")
plt.xlabel("Placement Status")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(y=df["AptitudeTestScore"])
plt.title("Aptitude Test Score Distribution")
plt.ylabel("Aptitude Test Score")
plt.show()
```

```python
sns.boxplot(x="PlacementStatus", y="CGPA", data=df, palette="coolwarm")
plt.title("CGPA vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("CGPA")
plt.show()
```

```python
sns.countplot(x="Internships", hue="PlacementStatus", data=df, palette="coolwarm")
plt.title("Internships vs Placement Status")
plt.xlabel("Number of Internships")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="PlacementTraining", hue="PlacementStatus", data=df, palette="coolwarm")
plt.title("Placement Training vs Placement Status")
plt.xlabel("Placement Training")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="SSC_Marks", y="CGPA", data=df, scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title("SSC Marks vs CGPA")
plt.xlabel("SSC Marks")
plt.ylabel("CGPA")
plt.show()

plt.figure(figsize=(8,6))
sns.regplot(x="HSC_Marks", y="CGPA", data=df, scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title("HSC Marks vs CGPA")
plt.xlabel("HSC Marks")
plt.ylabel("CGPA")
plt.show()
```

```python
df["CGPA_Range"] = pd.cut(df["CGPA"], bins=[5,6,7,8,9,10], labels=["5-6","6-7","7-8","8-9","9-10"])

plt.figure(figsize=(8,6))
sns.boxplot(x="CGPA_Range", y="Internships", data=df, palette="coolwarm")
plt.title("Internship Count Across CGPA Ranges")
plt.xlabel("CGPA Range")
plt.ylabel("Internships")
plt.show()
```

```python
features = ["CGPA", "AptitudeTestScore", "SoftSkillsRating", "Internships", "Projects"]
df["PlacementProbability"] = df[features].mean(axis=1)

sns.histplot(df["PlacementProbability"], kde=True, bins=20, color="blue")
plt.title("Estimated Placement Probability Distribution")
plt.xlabel("Placement Probability (Average of Key Factors)")
plt.ylabel("Frequency")
plt.show()
```
